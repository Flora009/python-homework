```python
from pathlib import Path
import csv

# Initialize menu_list to hold the contents of menu_data.csv

menu_list = []


# Set path and read the menu file
file_path = Path('./Resources/menu_data.csv')

with open(file_path,'r') as csvfile:
    csvreader = csv.reader(csvfile, delimiter=',')   
    header = next(csvreader)


    for row in csvreader:
        item = row[0]
        category = row[1]
        description = row[2]
        price = float(row[3])
        cost = float(row[4])
        
        # Store [key, value] pairs to menu_list which is a list of lists
        menu_list.append([item, category, description, price, cost])


# Initialize sales_list to hold the contents of menu_data.csv
sales_list = []

# Set path and read the sales data file
file_path = Path('./Resources/sales_data.csv')

with open(file_path,'r') as csvfile:
    csvreader = csv.reader(csvfile, delimiter=',')   
    header = next(csvreader)


    for row in csvreader:
        Line_Item = int(row[0])
        Date = row[1]
        Credit_Card_No = row[2]
        Quantity = int(row[3])
        Menu_Item = row[4]
        
        # Store [key, value] pairs to sales_list which is a list of lists
        sales_list.append([Line_Item, Date, Credit_Card_No, Quantity, Menu_Item])



```

```python

# Initialize dict object to hold our key-value pairs of items and metrics
# key is menu_item, value is report_metrics_header = [ '01-count','02-revenue','03-cogs','04-profit']

report = {}
report_metrics = {}
quantity = 0
menu_item = ''
rev = 0
cogs = 0
profit = 0

# Initialize a row counter variable
row_count = 0

# Loop over every row in the sales list object

for i in sales_list:
    if i[-1] not in report.keys():    
        menu_item = i[-1]
        quantity = int(i[3])
        report[menu_item] = {'01-count':quantity,'02-revenue':0,'03-cogs':0,'04-profit':0}

    elif i[-1] in report.keys():
        menu_item = i[-1]
        quantity = int(i[3] + report[menu_item]['01-count'])
        report[menu_item] = {'01-count':quantity,'02-revenue':0,'03-cogs':0,'04-profit':0}
        
for i in report.keys():
    for x in menu_list:
        if i == str(x[0]):
            item_price = float(x[3])
            item_cost = float(x[4])
            report[i]['02-revenue'] =  report[i]['01-count']*item_price
            report[i]['03-cogs'] = report[i]['01-count']*item_cost
            report[i]['04-profit'] = report[i]['02-revenue'] - report[i]['03-cogs']
        else:
            print(f"{i} does not equal {x}! NO MATCH!")
            
for items in report.items():
    print(items)

output_path = Path('./output_ramen.txt')
with open(output_path, 'w') as txtfile:
    for items in report.keys():
        txtfile.write(f'{items} {report[items]}\n')
```
